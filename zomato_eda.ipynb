## Zomato Exploratory Data Analysis


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('zomato.csv', encoding='latin-1')
dataset.head()
```

```python
dataset.columns
```

```python
dataset.shape
dataset.info()
dataset.describe()
```

```python
dataset.isnull().sum()
```

```python
[features for features in dataset.columns if dataset[features].isnull().sum() > 1]
```

```python
sns.heatmap(dataset.isnull(), yticklabels=False, cbar = False, cmap='viridis')
```

```python
data_country = pd.read_excel('Country-Code.xlsx')
data_country.head()
```

```python
final_dataset = pd.merge(dataset,data_country,on='Country Code', how='left')
final_dataset.head()
```

```python
final_dataset.info()
```

```python
final_dataset.columns
```

```python
country_names = final_dataset.Country.value_counts().index
country_names

```

```python
country_values = final_dataset.Country.value_counts().values
country_values
```

```python
plt.pie(country_values[:3], labels=country_names[:3],autopct='%1.2f%%')
```

```python
ratings = final_dataset.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', data = ratings)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating Count', hue = 'Rating color', data = ratings, palette=['black','red','orange','yellow','green','green'])
```

```python
sns.countplot(x = 'Rating color', data = ratings, palette=['black','red','orange','yellow','green','green'])
```
